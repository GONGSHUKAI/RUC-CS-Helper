#### 首先导入必要的工具包.

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
# 分词
import jieba
# 词云
from wordcloud import WordCloud,STOPWORDS
```

#### 导入三类影评的 DataFrame.

```python
# 好评
df_high = pd.read_csv('csv/WWY-h.csv', encoding = "utf-8", names = ['用户名', "短评"])
df_high
```

```python
# 中评
df_mid = pd.read_csv('csv/WWY-m.csv', encoding = "utf-8", names = ['用户名', "短评"])
df_mid
```

```python
# 差评
df_low = pd.read_csv('csv/WWY-l.csv', encoding = "utf-8", names = ['用户名', "短评"])
df_low
```

#### 影评分词函数

```python
# 排除一些代词, 口头词以及导演名的影响
# 这些词多见于介绍性句子, 而我们更关注评价性句子
# 比如说我们排除导演名、地名, 因为几乎每个评价都会涉及到, 我感觉对我们的分析意义不大
newstopwords = ['一部', '一起', '可能', '有点', '已经', '一波', 
                '都', '觉得', '我', '你', '他', '她', '刚看完', 
                '我会', '这座', '令人', '很', '应该', '真的', 
                '一种', '看来', '真是', '最后', '之前', '之后', 
                '不要', '那句', '没', '有', '没有', '一下', 
                '一个', '不了', '这种', '新海诚', '诚哥', '电影', 
                '人们', '不会', '知道', '这次', '这部', '东京', 
                '那种', '日本', '一点', '一次', '一句', '才', 
                '一面', '一往', '所有', '东西', '大概', '大约', 
                '两个', '整个', '这是', '看到']
# 分词函数
def word_cut (texts):
    words_list = []
    word_generator = jieba.cut(texts, cut_all=False)  # 返回的是一个迭代器
    # 停用词
    with open('tools/hit_stopwords.txt') as f:
        str_text = f.read()
    for word in word_generator:
        # 不是停用词文件内的且不是新加的停用词
        if word.strip() not in str_text and word.strip() not in newstopwords:
            words_list.append(word)
    return ' '.join(words_list)  # 注意是空格
```

#### 根据后面分词效果, 我们增加一些电影相关的词:

```python
# 一些在电影语境内有用的新词, 用于扩充分词词典
newwords = ['天气之子', '新海诚', '诚哥', '帆高', '阳菜', '中二',
            '君名', '你的名字', '青春剧', '男女主', '虎躯一震', '立民党',
            '傻缺', '双马尾', '爆笑感', '中二癌', '少年气', '众筹', 
            '冬雷震震', '夏雨雪', '乃敢与君绝', '天地合', '大鱼海棠', 
            '广告之子', '刚看完', '言叶', '言叶之庭', '糸守町', 
            '糸守', '山手线', '智熄']
for word in newwords:
    jieba.add_word(word)
```

```python
# 对短评列应用分词函数
df_high['短评(分词后)'] = df_high['短评'].apply(word_cut)
# 丢掉原来的短评
df_high_new = df_high.drop(['短评'], axis=1)
# 格式化输出到txt文件中
np.savetxt('WWY-h-parsed.txt', df_high_new.values, fmt = "[%s]:%s")
# 看看分词效果, 不断的往上面添加一些新词和新的停用词
df_high_new
```

```python
# 中评
df_mid['短评(分词后)'] = df_mid['短评'].apply(word_cut)
df_mid_new = df_mid.drop(['短评'], axis=1)
np.savetxt('WWY-m-parsed.txt', df_mid_new.values, fmt = "[%s]:%s")
df_mid_new
```

```python
# 差评
df_low['短评(分词后)'] = df_low['短评'].apply(word_cut)
df_low_new = df_low.drop(['短评'], axis=1)
np.savetxt('WWY-l-parsed.txt', df_low_new.values, fmt = "[%s]:%s")
df_low_new
```

#### 绘制词云的函数:

```python
# 标题使用中文
myfont = matplotlib.font_manager.FontProperties(fname="tools/HYQiHei-105.ttf", size=30)
matplotlib.rcParams['axes.unicode_minus'] = False

# 绘制词云
# words: 词列表; title: 词云标题; color: 背景色; mask: 轮廓图
def draw_wordcloud (words, title = None, color = 'white', mask = None):
    wordcloud = WordCloud(stopwords = STOPWORDS,
                          font_path="tools/HYQiHei-105.ttf",
                          background_color=color,
                          mask=mask,
                          width=1000,
                          height=1500
                         ).generate(words)
    plt.figure(1, figsize=(10, 15))
    # 标题中文
    plt.title(title, fontproperties = myfont)
    plt.imshow(wordcloud)
    plt.axis('off')
    # 保存图片
    plt.savefig(title+'.png')
    plt.show()
```

#### 分三类绘制词云:

```python
# 导入背景轮廓图用
import PIL .Image as image
# 收集词汇
words_h = ' '.join(df_high_new['短评(分词后)'])
print('好评词云:')
draw_wordcloud(words_h, title ='好评词云', color='white', mask=np.array(image.open('tools/mask1.jpg')))
```

### 好评词云分析:

首先观察几个最大的词, 分别是“世界”、“故事”、“少年”、“喜欢”、“你的名字”以及“画面”等等.

由此可见好评的用户应该是看过新海诚执导的电影的，属于电影的“铁粉”.

除此之外, 明显的词汇中看不到负面的用词, 整体评价积极.

```python
words_m = ' '.join(df_mid_new['短评(分词后)'])
print('中评词云:')
draw_wordcloud(words_m, title ='中评词云', color='white', mask=np.array(image.open('tools/mask1.jpg')))
```

### 中评词云分析:

和好评的词云有部分重叠, 比如“世界”、“故事”、“少年”和“你的名字”, 可见用户对电影有肯定之处;

同时“剧情”一词相比于好评词云大了许多, 有可能是这部分用户进行了客观的分析.

与此同时, 一些负面词汇开始出现, 比如“失望”、“不行”、“可惜”等, 总体而言是中偏上的评价.

```python
words_l = ' '.join(df_low_new['短评(分词后)'])
print('差评词云:')
draw_wordcloud(words_l, title ='差评词云', color='white', mask=np.array(image.open('tools/mask1.jpg')))
```

### 差评词云分析:
同样的, 还是有不少重叠, 如“故事”、“剧情”、“世界”等, 但态度估计不一定是正向的.

除此之外, “中二”一词比较大, 除此之外, 也有一些“幼稚”、“矫情”、“莫名其妙”等负向情感较明显的词汇, 

整体而言情绪是中偏下的.
